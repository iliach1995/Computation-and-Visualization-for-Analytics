```python
import pandas as pd
import numpy as np
import us
from pandas.api.types import is_numeric_dtype
```

```python
data = pd.read_csv ("PERM_Disclosure_Data_FY2022.csv", encoding = 'unicode_escape', engine ='python')
```

```python
print(data.columns)
```

```python
df = data[['CASE_STATUS', 'EMPLOYER_NAME', 'EMPLOYER_STATE_PROVINCE', 'EMPLOYER_NUM_EMPLOYEES', 'NAICS_CODE','PW_SOC_TITLE','EMP_CONTACT_STATE_PROVINCE','PW_WAGE','MINIMUM_EDUCATION']]
df.replace('', np.nan, inplace=True)
df.replace(0, np.nan, inplace=True)
df.replace('none', np.nan, inplace=True)
```

```python
df = df.dropna(how='any',axis=0) 
df.rename(columns = {'EMPLOYER_STATE_PROVINCE':'EMPLOYER_STATE','NAICS_CODE':'NAICS','PW_SOC_TITLE':'OCCUPATION','EMP_CONTACT_CITY':'EMP_CITY','EMP_CONTACT_STATE_PROVINCE':'EMP_STATE','PW_WAGE':'WAGE','MINIMUM_EDUCATION':'EDUCATION'}, inplace = True)
```

```python
tmp = df['CASE_STATUS']
tmp = pd.DataFrame(tmp);
tmp.rename(columns = {'CASE_STATUS': 'CASE_STATUS_NUMBER'}, inplace = True)
df = pd.concat([df, tmp], axis=1)
```

```python
df
```

```python
neworder = ['EMPLOYER_NAME','CASE_STATUS','CASE_STATUS_NUMBER','EMPLOYER_STATE','EMPLOYER_NUM_EMPLOYEES','NAICS','OCCUPATION','EMP_STATE','WAGE','EDUCATION']
df=df.reindex(columns=neworder)
```

```python
df
```

```python
Case_status_Number = {"Withdrawn": 0, "Denied": -1, "Certified-Expired":1, "Certified":1}
def ChangeCaseStatusNumber (x, Case_status_Number):
    return Case_status_Number.get(x)
df['CASE_STATUS_NUMBER'] = df['CASE_STATUS_NUMBER'].apply(lambda x : ChangeCaseStatusNumber(x, Case_status_Number))
df['CASE_STATUS'] = df['CASE_STATUS'].replace({'Certified-Expired':'Certified'}, regex = True)
```

```python
def ChangeState (x):
    state = us.states.lookup(x)
    return state
df['EMP_STATE'] = df['EMP_STATE'].apply(lambda x : ChangeState(x))
df['EMPLOYER_STATE'] = df['EMPLOYER_STATE'].apply(lambda x : ChangeState(x))
```

```python
df['WAGE'] = df['WAGE'].replace({'\$':''}, regex = True)
df['WAGE'] = df['WAGE'].replace({',':''}, regex = True)
df['WAGE'] = df['WAGE'].replace({'#############':'0'}, regex = True)
#############
```

```python
colnames = ['CASE_STATUS_NUMBER','EMPLOYER_NUM_EMPLOYEES','NAICS','WAGE']
for col_name in colnames:
        if not is_numeric_dtype(df[col_name]):
            #df[col_name]= df[col_name].astype(float)
            df[col_name]= df[col_name].apply(lambda x: x.replace("-", "")).astype(float)
```

```python
def WageChange (x):
    if x < 40000:
        return np.NaN
    else:
        return x
df['WAGE'] = df['WAGE'].apply(lambda x : WageChange(x))
```

```python
df
```

```python
NAICS_CODE_2022 = {
    11: "Agriculture, Forestry, Fishing, and Hunting",
    21: "Mining, Quarrying, and Oil and Gas Extraction",
    22: "Utilities", 
    23: "Construction",
    31: "Manufacturing", 
    32: "Manufacturing", 
    33: "Manufacturing",
    42: "Wholesale Trade", 
    44: "Retail Trade", 
    45: "Retail Trade",
    48: "Transportation and Warehousing", 
    49: "Transportation and Warehousing",
    51: "Information", 
    52: "Finance and Insurance",
    53: "Real Estate and Rental and Leasing", 
    54: "Professional, Scientific, and Technical Services", 
    55: "Management of Companies and Enterprises",
    56: "Administrative and Support and Waste Management and Remediation Services", 
    61: "Educational Services", 
    62: "Health Care and Social Assistance", 
    71: "Arts, Entertainment, and Recreation", 
    72: "Accommodation and Food Services",
    81: "Other Services (except PA)", 
    92: "Public Administration(PA)",
    99: "Unknown"}

def ChangeNAICS (x,NAICS_CODE_2022):
    a_string = str(x)
    a_length = len(a_string)
    c = int(a_string[0: 2])
    ans  = NAICS_CODE_2022.get(c)
    if ans == None:
        return np.NaN
    return NAICS_CODE_2022.get(c)

df['NAICS'] = df['NAICS'].apply(lambda x : ChangeNAICS(x,NAICS_CODE_2022))

```

```python
df
```

```python
keywords = ['Amazon', 'Walmart', 'Qualcomm' ,'Microsoft','Google','Intel','Tata Consultancy','NVIDIA','Oracle','Adobe','Apple','Twitter','Facebook']
def ChangeEmployerName (x):
    for keyword in keywords:
        if keyword.lower() in x.lower():
            return keyword
    return x

df['EMPLOYER_NAME'] = df['EMPLOYER_NAME'].apply(lambda x : ChangeEmployerName(x))
```

```python
keywords_edu = ['Master', 'Bachelor']
def ChangeEducation (x,keywords_edu):
    for keyword in keywords_edu:
        if keyword.lower() in x.lower():
            return keyword
    return x

df['EDUCATION'] = df['EDUCATION'].apply(lambda x : ChangeEducation(x,keywords_edu))

```

```python
df = df.dropna(how='any',axis=0) 
df
```

```python
df.to_csv('PERM2022CLEAN.csv', index = False)
```
