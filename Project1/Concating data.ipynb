```python
import pandas as pd
import numpy as np
data2017 = pd.read_csv ("PERM2017CLEAN.csv", encoding = 'unicode_escape', engine ='python')
temp = np.ones([data2017.shape[0],1])*2017
tmp = pd.DataFrame(temp);
tmp=tmp.astype('int')
tmp.rename(columns = {0: 'year'}, inplace = True)
data2017 = pd.concat([tmp, data2017], axis=1)

data_total = data2017
```

```python
filename = ["PERM2018CLEAN.csv","PERM2019CLEAN.csv","PERM2020CLEAN.csv","PERM2021CLEAN.csv","PERM2022CLEAN.csv"]
lst = [2018,2019,2020,2021,2022]
for i in range (5):
    data = pd.read_csv (filename[i], encoding = 'unicode_escape', engine ='python')
    temp = np.ones([data.shape[0],1])*lst[i]
    tmp = pd.DataFrame(temp);
    tmp=tmp.astype('int')
    tmp.rename(columns = {0: 'year'}, inplace = True)
    data = pd.concat([tmp, data], axis = 1)
    data_total = pd.concat([data_total, data])
```

```python
data_total['WAGE']=data_total['WAGE'].astype('int')

data_total
```

```python
df = data_total
tmp = df['CASE_STATUS']
tmp = pd.DataFrame(tmp);

temp2 = np.ones([tmp.shape[0],])
temp2 = pd.DataFrame({'CASE_STATUS_TOTAL': temp2})

data_total = pd.get_dummies(data_total, columns=['CASE_STATUS'])

df2 = pd.concat([data_total, tmp],axis = 1)
df2
```

```python
df2 = pd.concat([df2.reset_index(),temp2.reset_index()],axis = 1)
df2 = df2.drop('index', axis = 1)
```

```python
df2
```

```python
neworder =['year','EMPLOYER_NAME','CASE_STATUS','CASE_STATUS_TOTAL','CASE_STATUS_Certified','CASE_STATUS_Denied','CASE_STATUS_Withdrawn','EMPLOYER_STATE','EMPLOYER_NUM_EMPLOYEES','NAICS','OCCUPATION','EMP_STATE','WAGE','EDUCATION']
df2=df2.reindex(columns=neworder)
df2
```

```python
keywords = ['Netflix']
def ChangeEmployerName (x):
    for keyword in keywords:
        if keyword.lower() in x.lower():
            return keyword
    return x

df2['EMPLOYER_NAME'] = df2['EMPLOYER_NAME'].apply(lambda x : ChangeEmployerName(x))
```

```python
df2.to_csv('TotalPerm2017To2022.csv',index=False)
```
