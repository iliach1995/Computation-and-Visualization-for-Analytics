```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv ("TotalPerm2017To2022.csv", encoding = 'unicode_escape', engine ='python')
df
```

```python
def ChooseTopxRows(df, top_row_number):
    df_final = pd.DataFrame()
    for i in range (2017,2023):
        df_temp = df[df["year"] == i].head(top_row_number)
        df_final = pd.concat([df_final, df_temp])
    df_final = df_final.reset_index().drop(columns="index")
    return df_final
```

```python
df3 = df.copy()
df3 = df3.drop(['CASE_STATUS','CASE_STATUS_TOTAL','CASE_STATUS_Withdrawn','CASE_STATUS_Denied','EMPLOYER_NUM_EMPLOYEES','EMP_STATE','WAGE','EDUCATION','EMPLOYER_NAME'], axis = 1)
df3 = df3.groupby(by = ['year', 'EMPLOYER_STATE', 'NAICS','OCCUPATION'],as_index=False).sum()
industries = ['Professional, Scientific, and Technical Services', 'Information','Retail Trade', 'Manufacturing', 'Finance and Insurance']
df3_1 = pd.DataFrame()

for industry in industries:
        df_temp = df3[df3['NAICS'] == industry]
        df3_1 = pd.concat([df3_1, df_temp])

df3_2 = df3_1.groupby(by = ['year', 'EMPLOYER_STATE', 'NAICS', 'OCCUPATION'], as_index = False).sum()
df3_3 = df3_2.drop(['EMPLOYER_STATE', 'NAICS'],axis = 1)
df3_4 = df3_3.groupby(['year', 'OCCUPATION'], as_index = False).sum()
df3_5 = df3_4.sort_values(['year', 'CASE_STATUS_Certified'], ascending = [True, False])
df3_6 =  ChooseTopxRows(df3_5, 10)
df3_6['total accepted in 5 states'] = df3_6['year'].map(lambda x: 77068 if x == 2017 else (96157 if x == 2018 else (81135 if x == 2019 else (73006 if x == 2020 else (84993 if x == 2021 else (78478 if x == 2022 else ""))))))
df3_6['ratio'] = df3_6['CASE_STATUS_Certified']/df3_6['total accepted in 5 states']*100
df3_6.to_csv('Point 5 Occupation.csv')
df3_6
```
